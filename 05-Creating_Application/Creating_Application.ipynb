### Creating Reference Data CSVs for Columns
---

```python
# Import the Combined Data from /Wrangling_Data_Exports
wrangled_df = pd.read_csv("../04-Building_Model/Wrangling_Data_Exports/Wrangled_Data.csv")

# Display Dataframe
wrangled_df.head()
```

```python
# Create Aplphabetical Dataframes For Columns (Months, Times, Venues, Cities, States, Weather Conditions, Teams)
```

```python
# Create Dataframe for Cities

cities_list = wrangled_df["City"].unique()

# Create Dataframe of Unique Values
cities_df = pd.DataFrame(cities_list)

# Rename Column
cities_df=cities_df.rename(columns={0:"Cities"})

# Sort Alphabetically
cities_df = cities_df.sort_values("Cities", ascending=True)

# Reset Index
cities_df = cities_df.reset_index(drop=True)

# Export New Dataframe
cities_df.to_csv("Reference_Data/Cities_Data.csv")

# Display New Dataframe
cities_df
```

```python
# Create Dataframe for Months
months_list = wrangled_df["Month"].unique()

# Create Dataframe of Unique Values
months_df = pd.DataFrame(months_list)

# Rename Column
months_df=months_df.rename(columns={0:"Months"})

# Sort Alphabetically
months_df = months_df.sort_values("Months", ascending=True)

# Reset Index
months_df = months_df.reset_index(drop=True)

# Export New Dataframe
months_df.to_csv("Reference_Data/Months_Data.csv")

# Display New Dataframe
months_df
```

```python
# Create Dataframe for Times
times_list = wrangled_df["Time (EST)"].unique()

# Create Dataframe of Unique Values
times_df = pd.DataFrame(times_list)

# Rename Column
times_df=times_df.rename(columns={0:"Times"})

# Sort Alphabetically
times_df = times_df.sort_values("Times", ascending=True)

# Reset Index
times_df = times_df.reset_index(drop=True)

# Export New Dataframe
times_df.to_csv("Reference_Data/Times_Data.csv")

# Display New Dataframe
times_df
```

```python

# Create Dataframe for Venues
venues_list = wrangled_df["Venue"].unique()

# Create Dataframe of Unique Values
venues_df = pd.DataFrame(venues_list)

# Rename Column
venues_df=venues_df.rename(columns={0:"Venues"})

# Sort Alphabetically
venues_df = venues_df.sort_values("Venues", ascending=True)

# Reset Index
venues_df = venues_df.reset_index(drop=True)

# Export New Dataframe
venues_df.to_csv("Reference_Data/Venues_Data.csv")

# Display New Dataframe
venues_df
```

```python
# Create Dataframe for States

states_list = wrangled_df["State"].unique()

# Create Dataframe of Unique Values
states_df = pd.DataFrame(states_list)

# Rename Column
states_df=states_df.rename(columns={0:"States"})

# Sort Alphabetically
states_df = states_df.sort_values("States", ascending=True)

# Reset Index
states_df = states_df.reset_index(drop=True)

# Export New Dataframe
states_df.to_csv("Reference_Data/States_Data.csv")

# Display New Dataframe
states_df
```

```python
# Create Dataframe for Teams
teams_list = wrangled_df["Away Team"].unique()

# Create Dataframe of Unique Values
teams_df = pd.DataFrame(teams_list)

# Rename Column
teams_df=teams_df.rename(columns={0:"Teams"})

# Sort Alphabetically
teams_df = teams_df.sort_values("Teams", ascending=True)

# Reset Index
teams_df = teams_df.reset_index(drop=True)

# Export New Dataframe
teams_df.to_csv("Reference_Data/Teams_Data.csv")

# Display New Dataframe
teams_df
```

```python
# Create Dataframe for Weather

weather_conditions_list = wrangled_df["Weather Condition"].unique()

# Create Dataframe of Unique Values
weather_conditions_df = pd.DataFrame(weather_conditions_list)

# Rename Column
weather_conditions_df=weather_conditions_df.rename(columns={0:"Weather"})

# Sort Alphabetically
weather_conditions_df = weather_conditions_df.sort_values("Weather", ascending=True)

# Reset Index
weather_conditions_df = weather_conditions_df.reset_index(drop=True)

# Export New Dataframe
weather_conditions_df.to_csv("Reference_Data/Weather_Data.csv")

# Display New Dataframe
weather_conditions_df
```
